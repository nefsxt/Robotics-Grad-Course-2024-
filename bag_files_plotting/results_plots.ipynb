```python
from bagpy import bagreader
import bagpy
import pandas as pd
import seaborn as sea
import matplotlib.pyplot as plt
```

```python
b = bagreader(r'C:\Users\Nefeli\Desktop\robotics_final\omadang.bag')
```

```python
# Read Velocity Data
ms = b.vel_data()
vel = pd.read_csv(ms[0])
```

```python
# Read odometry Data
odom = b.odometry_data()
odomdata = pd.read_csv(odom[1])
```

```python
odomdata
```

```python
plt.plot(odomdata['pose.x'], odomdata['pose.y'], marker='o')
plt.xlabel('pose.x')
plt.ylabel('pose.y')
plt.title('Trajectory ')
plt.show()
plt.savefig('trajectory.png')
```

```python
plt.plot(vel['Time'],vel['linear.x'], marker='o')
plt.ylabel('linear.x')
plt.xlabel('Time (s)')
plt.title('Linear Velocity X  ')
plt.show()
plt.savefig('vel_lin_x.png')
```

```python
plt.plot(vel['Time'],vel['angular.z'], marker='o')
plt.ylabel('angular.z')
plt.xlabel('Time (s)')
plt.title('Angular Velocity Z  ')
plt.show()
plt.savefig('vel_ang_z.png')
```
